## Spheres

```python
from pyNanoMatBuilder import crystalNPs as cNP

vID.centerTitle(f"Ru Sphere")
RuNP = cNP.Crystal("Ru",size=[2])
RuNP = RuNP.makeNP()

write("coords/SphericalRuNP.xyz", RuNP)

pnmbu.moi(RuNP)
pnmbu.MolSym(RuNP)
```

## Ellipsoids

```python
from pyNanoMatBuilder import crystalNPs as cNP

vID.centerTitle(f"Ru Ellipsoid")
RuNP = cNP.Crystal("Ru",shape='ellipsoid',size=[1,1,2])
RuNP = RuNP.makeNP()

write("coords/EllipsoidalRuNP.xyz", RuNP)

pnmbu.moi(RuNP)
pnmbu.MolSym(RuNP)
```

### Regular icosahedra

```python
from pyNanoMatBuilder import platonicNPs as pNP

nshells = 3
vID.centerTitle(f"Ico {nshells} shells")
ico = pNP.regIco("Au",2.7,nshells)

vID.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
ico.prop()

vID.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
aseico,atico = ico.coords()
view(aseico)
write("coords/ico.xyz", aseico)

pnmbu.moi(aseico)
pnmbu.MolSym(aseico)
```

# Geometry optimization with ase

## Regular fcc octahedra

```python
from pyNanoMatBuilder import platonicNPs as pNP
fccOh = pNP.regfccOh("Au",2.7,5)
aseOh,atOh = fccOh.coords()
view(aseOh)
aseOhOpt = pnmbu.optimizeEMT(aseOh,'./coords/fccOh')

pnmbu.moi(aseOhOpt)
pnmbu.MolSym(aseOhOpt)
```

## Regular dodecahedra

```python
from pyNanoMatBuilder import platonicNPs as pNP
rdd = pNP.regDD("Au",2.7,4)
aserdd,atrdd = rdd.coords()
view(aserdd)
aserddOpt = pnmbu.optimizeEMT(aserdd,'./coords/rdd')

pnmbu.moi(aserddOpt)
pnmbu.MolSym(aserddOpt)
```
