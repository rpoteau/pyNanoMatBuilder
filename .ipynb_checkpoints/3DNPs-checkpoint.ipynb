# 3D NPs. Generation of coordinates

## Initialization

```python
import numpy as np
from ase.io import write
from ase import Atoms
from ase.visualize import view

import pyPolyhedra.utils as ppu
```

## Platonic solids

### Regular icosahedra

```python
# from ase.cluster import Icosahedron

# atoms = Icosahedron("Au", 8, 4.1536)
# atoms.center(vacuum=10)

# write("coords/ico.xyz", atoms)
# write("coords/ico.cif", atoms)
```

```python
from pyPolyhedra import platonicNPs as pNP

nshells = 1
t4pPC.centerTitle(f"Ico {nshells} shells")
ico1 = pNP.regIco(2.7,nshells)

t4pPC.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
ico1.prop()

t4pPC.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c1,i1 = ico1.coords()
print(ico1.nAtoms)
print(ico1.nAtomsPerShell)
ico1 = Atoms('Au'*ico1.nAtoms, positions=c1)
view(ico1)
write("coords/ico1.xyz", ico1)

t4pPC.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
ppu.moi(ico1)
print()

#######################################################################################################
nshells = 10
t4pPC.centerTitle(f"IC {nshells} shells")
IC2 = pNP.regIco(2.7,nshells)

t4pPC.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
IC2.prop()

t4pPC.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c2,i2 = IC2.coords()
print(IC2.nAtoms)
print(IC2.nAtomsPerShell)
ico2 = Atoms('Au'*IC2.nAtoms, positions=c2)
view(ico2)
write("coords/ico2.xyz", ico2)

t4pPC.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
ppu.moi(ico2)
```

### Regular fcc tetrahedra

<div class="intro">

[Wikipedia](https://en.wikipedia.org/wiki/Tetrahedron)
</div>

```python
from pyPolyhedra import platonicNPs as pNP
nLayers = 9
t4pPC.centerTitle(f"fcc Td: {nLayers} atoms/edge = number of layers")
Td = pNP.regfccTd(2.7,nLayers)

t4pPC.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
Td.prop()

t4pPC.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c1,i1 = Td.coords()
print(f"Total number of atoms = {Td.nAtoms:.0f}")
print(f"Number of atoms per layer = {Td.nAtomsPerLayerAnalytic()}")
Td = Atoms('Au'*Td.nAtoms, positions=c1)
write("coords/Td.xyz", Td)
view(Td)

t4pPC.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
ppu.moi(Td)
```
