```python
##################
# Initialization #
##################

import os
import sys

print(os.getcwd())
cwd0 = './styles/'
sys.path.append(cwd0)

import visualID as vID
from visualID import  fg, hl, bg
vID.init(cwd0)

import numpy as np
from ase.io import write
from ase import Atoms
from ase.visualize import view

import pyNanoMatBuilder.utils as pnmbu
```

# Polyhedra

## Platonic solids

### Regular icosahedra

```python
from pyNanoMatBuilder import platonicNPs as pNP

nshells = 1
vID.centerTitle(f"Ico {nshells} shells")
ico1 = pNP.regIco(2.7,nshells)

vID.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
ico1.prop()

vID.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c1,i1 = ico1.coords()
print(ico1.nAtoms)
print(ico1.nAtomsPerShell)
ico1 = Atoms('Au'*ico1.nAtoms, positions=c1)
view(ico1)
write("coords/ico1.xyz", ico1)

vID.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
pnmbu.moi(ico1)
print()

#######################################################################################################
nshells = 10
vID.centerTitle(f"IC {nshells} shells")
IC2 = pNP.regIco(2.7,nshells)

vID.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
IC2.prop()

vID.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c2,i2 = IC2.coords()
print(IC2.nAtoms)
print(IC2.nAtomsPerShell)
ico2 = Atoms('Au'*IC2.nAtoms, positions=c2)
view(ico2)
write("coords/ico2.xyz", ico2)

vID.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
pnmbu.moi(ico2)
```

### Regular fcc tetrahedra

<div class="intro">

[Wikipedia](https://en.wikipedia.org/wiki/Tetrahedron)
</div>

```python
from pyNanoMatBuilder import platonicNPs as pNP
nLayers = 9
vID.centerTitle(f"fcc Td: {nLayers} atoms/edge = number of layers")
Td = pNP.regfccTd(2.7,nLayers)

vID.centertxt("Properties",bgc='#007a7a',size='14',weight='bold')
Td.prop()

vID.centertxt("Generation of coordinates",bgc='#007a7a',size='14',weight='bold')
c1,i1 = Td.coords()
print(f"Total number of atoms = {Td.nAtoms:.0f}")
print(f"Number of atoms per layer = {Td.nAtomsPerLayerAnalytic()}")
Td = Atoms('Au'*Td.nAtoms, positions=c1)
write("coords/Td.xyz", Td)
view(Td)

vID.centertxt("Moments of inertia",bgc='#007a7a',size='14',weight='bold')
pnmbu.moi(Td)
```
