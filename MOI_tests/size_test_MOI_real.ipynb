## Difference between the real sizes (measured on JMOL) and the sizes computed from MOI

```python
#data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('diff in Angs :')

size_MOI_sphere=np.array([1.98, 3.99, 5.99, 8.02, 9.99,20.01])
size_JMOL_sphere=np.array([1.85, 3.88, 5.92, 7.96, 9.86,19.78])
diff_sphere=size_MOI_sphere-size_JMOL_sphere
pourcentage_sphere=100-(size_JMOL_sphere/size_MOI_sphere)*100
print('diff sphere=',diff_sphere, ', pourcentage=',pourcentage_sphere)

size_MOI_cube=np.array([2.04, 4.05, 6.24, 8.27, 10.60,21.05])
size_JMOL_cube=np.array([1.87, 3.90, 6.08, 8.11, 10.44,20.89])
diff_cube=size_MOI_cube-size_JMOL_cube
pourcentage_cube=100-(size_JMOL_cube/size_MOI_cube)*100
print('diff cube=',diff_cube, ', pourcentage=',pourcentage_cube)

size_MOI_wire=np.array([1.83,3.87,5.91,7.95,9.99,20.18])
size_JMOL_wire=np.array([1.63,3.67,5.71,7.75,9.79,19.98])
diff_wire=size_MOI_wire-size_JMOL_wire
pourcentage_wire=100-(size_JMOL_wire/size_MOI_wire)*100
print('diff wire=',diff_wire, ', pourcentage=',pourcentage_wire)


size_MOI_ellipsoid=np.array([1.98,3.97,6.02,8.06,9.94,19.89])
size_JMOL_ellipsoid=np.array([1.85,3.88,5.92,7.57,9.59,19.78])
diff_ellipsoid=size_MOI_ellipsoid-size_JMOL_ellipsoid
pourcentage_ellipsoid=100-(size_JMOL_ellipsoid/size_MOI_ellipsoid)*100
print('diff ellipsoid=',diff_ellipsoid, ', pourcentage=',pourcentage_ellipsoid)


size_MOI_parallepiped=np.array([1.83,3.86,5.92,7.94,10,20.18])
size_JMOL_parallepiped=np.array([1.63,3.67,5.71,7.75,9.79,20])
diff_parallepiped=size_MOI_parallepiped-size_JMOL_parallepiped
pourcentage_parallepiped=100-(size_JMOL_parallepiped/size_MOI_parallepiped)*100
print('diff parallepiped=',diff_parallepiped, ', pourcentage=',pourcentage_parallepiped)


size_MOI_octa=np.array([2.27,4.19,6.10,8.01,9.92,20.24])
size_JMOL_octa=np.array([1.90,3.89,5.72,7.64,9.54,19.86])
diff_octa=size_MOI_octa-size_JMOL_octa
pourcentage_octa=100-(size_JMOL_octa/size_MOI_octa)*100
print('diff octa=',diff_octa, ', pourcentage=',pourcentage_octa)

size_MOI_dd=np.array([2.31,4.88,8.01,10.87,12.60,25.46])
size_JMOL_dd=np.array([2.29,5.15,7.46,10.03,13.73,28.04 ])
diff_dd=abs(size_MOI_dd-size_JMOL_dd)
pourcentage_dd=abs(100-(size_JMOL_dd/size_MOI_dd)*100)
print('diff dd=',diff_dd, ', pourcentage=',pourcentage_dd)





```

```python
%matplotlib ipympl
plt.plot(size_MOI_cube,size_JMOL_cube,color='g',label=['cube'])
plt.plot(size_MOI_sphere,size_JMOL_sphere,color='b',label=['sphere'])
plt.plot(size_MOI_wire,size_JMOL_wire,color='y',label=['wire'])
plt.plot(size_MOI_ellipsoid,size_JMOL_ellipsoid,color='m',label=['ellipsoid'])
plt.plot(size_MOI_parallepiped,size_JMOL_parallepiped,color= 'black' ,label=['parallepiped'])
plt.plot(size_MOI_octa,size_JMOL_octa,color= 'orange' ,label=['octahedron'])
plt.plot(size_MOI_dd,size_JMOL_dd,color= 'grey' ,label=['dodecahedron'])
plt.plot(np.linspace(2,20,19),np.linspace(2,20,19),color='r',marker='o')
plt.xlabel('size from MOI')
plt.ylabel('size from JMOL')
plt.title('size from JMOL = f(size from MOI)')
plt.legend()
```

```python
%matplotlib ipympl
p=[2,4,6,8,10,20]
fig, ax = plt.subplots()
plt.plot(p,diff_cube,marker='o',color='g',label=['cube'])
plt.plot(p,diff_sphere,marker='o',color='b',label=['sphere'])
plt.plot(p,diff_wire,marker='o',color='y',label=['wire'])
plt.plot(p,diff_ellipsoid,marker='o',color='m',label=['ellipsoid'])
plt.plot(p,diff_parallepiped,marker='o',color='black',label=['parallepiped'])
plt.plot(p,diff_octa,marker='o',color='orange',label=['octahedron'])
plt.plot(p,diff_dd,marker='o',color='grey',label=['dodecahedron'])
plt.xlabel('different sizes in nm (approximated)')
plt.ylabel('error in nm')
plt.title('error in nm for different sizes')
plt.legend()
```

```python
%matplotlib ipympl
fig, ax = plt.subplots()
p=[2,4,6,8,10,20]
plt.plot(p,pourcentage_cube,marker='o',color='g',label=['cube'])
plt.plot(p,pourcentage_sphere,marker='o',color='b',label=['sphere'])
plt.plot(p,pourcentage_wire,marker='o',color='y',label=['wire'])
plt.plot(p,pourcentage_ellipsoid,marker='o',color='m',label=['ellipsoid'])
plt.plot(p,pourcentage_parallepiped,marker='o',color='black',label=['parallepiped'])
plt.plot(p,pourcentage_octa,marker='o',color='orange',label=['octahedron'])
plt.plot(p,pourcentage_dd,marker='o',color='grey',label=['dodecahedron'])
plt.xlabel('different sizes in nm (approximated)')
plt.ylabel('pourcentage')
plt.title('pourcentages error')
plt.legend()

```
