### Defining the wire ourselves :
- direction = orientation of the wire
- nRot = number of sides
- refPlane = plane that is rotated to create the wire


#### [0,0,1] RuNW with 6 sides, without pbc (periodic boundaries) and refPlane = (1 0 0)


```python

importlib.reload(cyNP)
importlib.reload(pyNMBu)

direction = [0,0,1] #direction of the nanoprism's axis
dim = [2,5] #diameter and length
refPlane = [1,0,0] #refPlane = plane to rotate
nRot = 6 #6 plans 
RuWire = cyNP.Crystal("Ru hcp",shape='wire',size=dim,directionWire=direction,nRotWire=nRot,refPlaneWire=refPlane,pbc=False,aseView= False) #planeRotation pour générer des plans symétriques autour de l’axe du wire


pyNMBu.writexyz("coords/RuWire.xyz", RuWire.NP)
with open("coords/RuWire.script", 'w') as f: f.write(RuWire.jMolCS)

RuWire.NP.center()
write("coords/RuWire.cif", RuWire.NP)
RuWire.NPcs.center()
write("coords/RuWire_CoreSurface.cif", RuWire.NPcs)
```
